```python
class Node(object):
    def __init__(self, i, x, y):
        self.i = i
        self.x = x
        self.y = y
```

```python
for i, node in enumerate(G.nodes()):
    G.node[node]["object"] = Node(i, node[0], node[1])
```

```python
class Container(object):
    def __init__(self, n):
        self.elements = [Element(i, self) for i in range(n)]
        self.element_dict= {i: self.elements[i] for i in range(n)}
    
class Element(object):
    def __init__(self, i, container):
        self.i = i
        self.container = container
```

```python
c = Container(5)
```
