```python
from TributeModel import Model

import json
import random

import networkx as nx
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
#G = nx.grid_2d_graph(10,10)
G = nx.watts_strogatz_graph(10, 3, 0.2)
#G = nx.scale_free_graph(10)
#G = G.to_undirected()
#G = nx.erdos_renyi_graph(10, 1)

G = nx.relabel_nodes(G, {tup: i for i, tup in enumerate(G.nodes())})
m = Model(G)
m.verbose = False
m.heuristic = True
m.random_execution = False
m.tribute = 100
m.max_depth = 2
#m.actions_per_turn = 10
pos = nx.spring_layout(G)
```

```python
for agent in m.agents.values():
    agent.max_depth = random.randint(10, 20)
```

```python
#m.verbose = False
for i in range(300):
    m.step()
    #print i
    #print "*\t*\t*"
    if (i % 30) == 0:
        print i
```

```python
nx.draw(m.graph, pos=pos, node_size=[m.agents[i].wealth for i in m.agents])
```
