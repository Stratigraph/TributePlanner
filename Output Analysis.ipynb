## Advantage of lookahead in heterogenous model

Load and combine the data from several different runs, for a total of 50 runs, such that each run has a different graph. Agent lookahead varies between 0 and 5. Run some statistics to check whether the distributions of final wealth differ by lookahead, and then run a regression to figure out the effect.

```python
with open("batch1.json") as f:
    data_1 = json.load(f)
    
with open("batch3.json") as f:
    data_2 = json.load(f)

with open("batch4.json") as f:
    data_3 = json.load(f)
```

```python
raw_data = data_1 + data_2 + data_3
```

```python
records = []
for i, model in enumerate(raw_data):
    G = nx.from_dict_of_dicts(model["graph"])
    degrees = G.degree()
    for id_num, agent in model["agents"].items():
        agent["degree"] = degrees[id_num]
        agent["iteration"] = i
        records.append(agent)
```

```python
agent_data = pandas.DataFrame(records)
```

```python
agent_data.head()
```

```python
agent_data.corr()
```

```python
plt.scatter(agent_data.wealth, agent_data.max_depth)
```

```python
d0 = agent_data.wealth[agent_data.max_depth==0]
d1 = agent_data.wealth[agent_data.max_depth==1]
d2 = agent_data.wealth[agent_data.max_depth==2]
d3 = agent_data.wealth[agent_data.max_depth==3]
d4 = agent_data.wealth[agent_data.max_depth==4]
d5 = agent_data.wealth[agent_data.max_depth==5]
```

```python
scipy.stats.ttest_ind(d0, d4)
```

```python
model = sm.OLS(agent_data.wealth, agent_data[["degree", "max_depth", "starting_wealth"]])
fit = model.fit()
print fit.summary()
```

What if we drop the peripheral agents?

```python
winning_agents = agent_data[agent_data.wealth > 500]
```

```python
model = sm.OLS(winning_agents.wealth, winning_agents[["degree", "max_depth", "starting_wealth"]])
fit = model.fit()
print fit.summary()
```

## Distribution of outcomes and wars with heuristic, zero-lookahead, and heterogenous lookahead

```python
with open("batch6.json") as f:
    heuristic_data = json.load(f)

with open("batch7.json") as f:
    nolook_data = json.load(f)
```

```python
final_wealth_1 = []
for model in heuristic_data:
    final_wealth_1 += [agent["wealth"] for agent in model["agents"].values()]
plt.hist(final_wealth_1)
```

```python
final_wealth_2 = []
for model in nolook_data:
    final_wealth_2 += [agent["wealth"] for agent in model["agents"].values()]
plt.hist(final_wealth_2)
```

```python
final_wealth_3 = []
for model in raw_data:
    final_wealth_3 += [agent["wealth"] for agent in model["agents"].values()]
plt.hist(final_wealth_3)
```

```python
final_wealth_4 = []
for model in data_4:
    final_wealth_4 += [agent["wealth"] for agent in model["agents"].values()]
plt.hist(final_wealth_4)
```

### Check whether these are drawn from the same distribution?

```python
wealth_dists = [final_wealth_1, final_wealth_2, final_wealth_3, final_wealth_4]
```

```python
for s1, s2 in combinations(wealth_dists, 2):
    print scipy.stats.mannwhitneyu(s1, s2)
```

```python
for s1, s2 in combinations(wealth_dists, 2):
    print scipy.stats.ks_2samp(s1, s2)
```
